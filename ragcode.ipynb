```python
# Quantization Configuration for LLaMA model
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4"
)

# Load LLaMA Model (quantized version)
model = AutoModelForCausalLM.from_pretrained(
    'meta-llama/Meta-Llama-3-8B',
    quantization_config=quantization_config
)
tokenizer = AutoTokenizer.from_pretrained('meta-llama/Meta-Llama-3-8B')

# Move model to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
# Step 6: Retrieval and Generation
query = "What is gradient descent?"  # Example query
retrieved_docs = retriever.get_relevant_documents(query)
context = " ".join([doc.page_content for doc in retrieved_docs])

# Combine query and context for input
input_text = f"Context: {context}\n\nQuery: {query}"
inputs = tokenizer(input_text, return_tensors="pt", truncation=True).to(device)

# Generate Response
output = model.generate(inputs['input_ids'], max_length=200, num_beams=5, early_stopping=True)
response = tokenizer.decode(output[0], skip_special_tokens=True)

# Step 7: Output Response
print("Response:")
print(response)
```
